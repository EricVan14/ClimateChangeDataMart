```python
import pandas as pd
from sklearn.svm import OneClassSVM
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('/Users/yuwenbo/Downloads/test/Stagged_data.csv')
```

```python
#Drop the columns contain words and bucket.
df = df.drop(columns=['Country'])
df = df.drop(columns=['Emissions Unit'])
df = df.drop(columns=['ColdestMonth'])
df = df.drop(columns=['HottestMonth'])
df = df.drop(columns=['CO2_Emissions_Bucket'])
df = df.drop(columns=['GreenhouseGas_Emissions_Bucket'])
df = df.drop(columns=['Methane_Emissions_Bucket'])
df = df.drop(columns=['NitrousOxide_Emissions_Bucket'])
df = df.drop(columns=['FluorinatedGases_Emissions_Bucket'])
df = df.drop(columns=['Total_emissions_Bucket'])
df = df.drop(columns=['YearlyTemperatureChange_Bucket'])
```

```python
#Select features for outlier detection
features = df.columns 


scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[features])

#Detecting outliers using One-Class SVM
ocsvm = OneClassSVM(gamma='auto', nu=0.01)
ocsvm.fit(scaled_features)

#Predict the outliers, where -1 indicates an outlier and 1 indicates a normal observation
df['Outlier'] = ocsvm.predict(scaled_features)

df
```
