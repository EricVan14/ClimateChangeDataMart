```python
import pandas as pd

df= pd.read_csv('Stagged_data.csv')
df.head()
```

```python
# Drill down to see CO2 emissions for "Botswana" by year
botswana_emissions_by_year = df[df['Country'] == 'Botswana'][['Year', 'CO2_Emissions']].sort_values('Year')
print(botswana_emissions_by_year)
```

```python
# Roll up to see total CO2 emissions
total_co2_emissions = df['CO2_Emissions'].sum()
print(total_co2_emissions)
```

```python
# Slice to get all data for the year 2020
data_2020 = df[df['Year'] == 2020]
print(data_2020)
```

```python
# Dice to get data for "Grenada" from 1970 to 1974
grenada_1970_1974 = df[(df['Country'] == 'Grenada') & (df['Year'].between(1970, 1974))]
print(grenada_1970_1974)
```

```python
# Selecting rows for 'Lebanon' and 'Tonga' from 1980 to 1990
selected_countries = ['Lebanon', 'Tonga']
selected_years = range(1980, 1991) 

# Dice operation to create a sub-cube
dice_sub_cube = df[
    (df['Country'].isin(selected_countries)) &
    (df['Year'].isin(selected_years))
][['Country', 'Year', 'CO2_Emissions', 'Methane_Emissions']]

print(dice_sub_cube)
```

```python
#Time Period Analysis: Summarize CO2 emissions by year
emissions_by_year = df.groupby('Year')['CO2_Emissions'].sum().reset_index()
print(emissions_by_year)
```

```python
#Country-Specific Analysis: Analyze emissions for specific countries over a range of years
specific_countries = df[df['Country'].isin(['Ukraine', 'Norway'])]
emissions_by_country_and_year = specific_countries.groupby(['Country', 'Year'])['CO2_Emissions'].sum().reset_index()
print(emissions_by_country_and_year)
```

```python
#Temperature Analysis: Investigate the average yearly temperature change
average_temp_change = df.groupby('Year')['YearlyTemperatureChange'].mean().reset_index()
print(average_temp_change)
```

```python
#Emissions and Temperature Analysis: Explore the relationship between total emissions and yearly temperature change
emissions_temp_relation = df.groupby('Year').agg({'Total_emissions': 'sum', 'YearlyTemperatureChange': 'mean'}).reset_index()
print(emissions_temp_relation)
```

```python
##Iceberg Queries

# Find the five years with the highest CO2 emissions growth:
df['CO2_Emissions_Growth'] = df.sort_values(by=['Year'])['CO2_Emissions'].diff()
top_growth_years = df.groupby('Year')['CO2_Emissions_Growth'].sum().nlargest(5).reset_index()
print(top_growth_years)

# Find the five countries with the highest decrease in methane emissions over the dataset's span:
initial_methane = df.groupby('Country')['Methane_Emissions'].first().reset_index(name='Initial_Methane')
final_methane = df.groupby('Country')['Methane_Emissions'].last().reset_index(name='Final_Methane')
methane_change = pd.merge(initial_methane, final_methane, on='Country')
methane_change['Change'] = methane_change['Final_Methane'] - methane_change['Initial_Methane']
top_decreases = methane_change.sort_values(by='Change').head(5)
print(top_decreases)
```

```python
## Windowing Queries
#Year-over-year change in total emissions for each country:
df['Total_Emissions'] = df['CO2_Emissions'] + df['GreenhouseGas_Emissions'] + df['Methane_Emissions'] + df['NitrousOxide_Emissions'] + df['FluorinatedGases_Emissions']
df.sort_values(by=['Country', 'Year'], inplace=True)
df['YoY_Emissions_Change'] = df.groupby('Country')['Total_Emissions'].pct_change() * 100
print(df[['Country', 'Year', 'YoY_Emissions_Change']])
```

```python
##Using the Window Clause Equivalent (no window clause like SQL so have to use .shift())
#Compare CO2 emissions to the previous year for each country:
df['Previous_Year_CO2'] = df.groupby('Country')['CO2_Emissions'].shift(1)
df['CO2_Change_From_Prev_Year'] = df['CO2_Emissions'] - df['Previous_Year_CO2']
print(df[['Country', 'Year', 'CO2_Emissions', 'Previous_Year_CO2', 'CO2_Change_From_Prev_Year']])
```
